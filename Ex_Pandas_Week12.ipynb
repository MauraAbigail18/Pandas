```python
#Seleksi weekdays dan weekends pada data deret waktu

#PERSIAPAN DATA FRAME

n_rows = 365
n_cols = 2 

cols = ['col1','col2']

df=pd.DataFrame(np.random.randint(1,20, size=(n_rows,n_cols)), columns = cols)

df.index = pd.util.testing.makeDateIndex(n_rows, freq = 'D')
df
```

```python
weekdays_df = df[df.index.dayofweek.isin([0,1,2,3,4])] #seleksi data yang termasuk ke dalam weekdays sebanyak 7 data
weekdays_df.head(7)
```

```python
weekends_df = df[df.index.dayofweek.isin([5,6])] #seleksi data yang termasuk ke dalam weekends sebanyak 7 data
weekends_df.head(7)
```

```python
#Penanganan kolom dengan tipe data beragam

#PERSIAPAN DATA FRAME

d = {'nama':['bejo','tejo','wati','tiwi','cecep'],
     'ipk':[2,'3',3, 2.75,'3.25']}
df = pd.DataFrame(d)
df
```

```python
#Deteksi dan penanganan kolom dengan tipe data beragam (mixed data types)

df.dtypes
```

```python
df['ipk'].apply(type) #mendeteksi keberadaan mixed data types pada kolom ipk
```

```python
df['ipk'].apply(type).value_counts() #mengetahui berapa jumlah keberadaan mixed data types pada kolom ipk
```

```python
df['ipk'] = df['ipk'].astype(float) #menyeragamkan tipe data di kolom ipk yang semula mixed data types menjadi tipe data float
```

```python
df['ipk'].apply(type).value_counts() #mengetahui berapa jumlah keberadaan tipe data float pada kolom ipk yang sudah diseragamkan tadi
```

```python
#Cummulative Count pada Pandas Data Frame

#PERSIAPAN DATA FRAME

d = {'penjual':['bejo','tejo','wati','bejo','cecep','tejo','wati','bejo'],
     'barang':['monitor','monitor','keyboard', 'mouse','keyboard','monitor','laptop','monitor']}
df = pd.DataFrame(d)
df
```

```python
#Mengenal Cummulative Count dengan cumcount()

df['count_tiap_penjual'] = df.groupby('penjual').cumcount() + 1 #penjumlahan nilai pada kolom penjual
df
```

```python
df['count_tiap_barang'] = df.groupby('barang').cumcount() + 1 #penjumlahan nilai pada kolom barang
df
```

```python
df['count_pasangan_kolom'] = df.groupby(['penjual', 'barang']).cumcount() + 1 #penjumlahan nilai pada kolom penjual dan barang
df
```
