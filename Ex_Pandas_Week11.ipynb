```python
#IMPORT MODULES

import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

```python
#Agregasi pada banyak kolom Data Frame dengan agg

#PERSIAPAN DATA FRAME

df = pd.read_csv('/content/titanicfull.csv') #Sebagai Data Frame
df.head()
```

```python
#Menerapkan agregasi untuk sejumlah kolom dengan agg() secara bersamaan
df.groupby('pclass').agg({'pclass':'count',
                          'age':['mean','max'],
                          'survived':'mean'})
```

```python
#Menerapkan agregasi untuk sejumlah kolom dengan agg() secara bersamaan dengan tampilan yang lebih rapi
df.groupby('pclass').agg(n_pass=('pclass','count'),
                          avg_age=('age','mean'),
                          max_age=('age','max'),
                          survival_rate=('survived','mean'))
```

```python
#Menangani whitespace pada Pandas Data Frame

#PERSIAPAN DATA FRAME

data = {'nim':['10','11','12','13','  '],
        'nama':['adi','  ','tejo','  ','bejo']}
df = pd.DataFrame(data)
df
```
