```python
#Resampling pada data deret waktu (time series data)
#IMPORT MODULES

import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

```python
#PERSIAPAN DATA FRAME
n_rows = 365*24 #365 merepresentasikan jumlah hari dalam 1 tahun, 24 merepresenrasikan jumlah jam dalam 1 hari
n_cols = 2
cols = ['col 1', 'col 2']
df=pd.DataFrame(np.random.randint(1,20, size=(n_rows,n_cols)), columns = cols)

df.index = pd.util.testing.makeDateIndex(n_rows, freq = 'H')
df
```

```python
#RESAMPLING DATA DENGAN INTERVAL MONTHLY
df.resample('M')['col 1'].sum().to_frame() #Untuk melakukan resampling berdasarkan bulan
```

```python
#RESAMPLING DATA DENGAN INTERVAL DAILY
df.resample('D')['col 1'].sum().to_frame() #Untuk melakukan resampling berdasarkan hari
```

```python
#PERSIAPAN DATA FRAME

n_rows =5
n_cols= 2

cols = ['omset', 'operasional']
df=pd.DataFrame(np.random.randint(1,20, size=(n_rows,n_cols)), columns = cols)

df
```

```python
df['omset'] = df['omset'] *100_000
df['operasional'] = df['operasional'] *10_000
df
```

```python
df.index = pd.util.testing.makeDateIndex(n_rows, freq = 'D')
df = df.reset_index() #Untuk menjadikan indeks saat ini yaitu indeks date time menjadi sebuah kolom biasa dengan nama indeks
df = df.rename(columns = {'index':'tanggal'})
df
```

```python
#MELAKUKAN FORMATTING TAMPILAN DATA FRAME

formatku = {'tanggal' : '{:%d/%m/%y}', #format hari, bulan, dan tahun
            'operasional' : 'Rp{:.2f}', #untuk meyertakan 2 digit ekstra di belakang koma
            'omset' : 'Rp{:.2f}'}
laporan = df.style.format(formatku)
laporan#MELAKUKAN FORMATTING TAMPILAN DATA FRAME

formatku = {'tanggal' : '{:%d/%m/%y}', #format hari, bulan, dan tahun
            'operasional' : 'Rp{:.2f}', #untuk meyertakan 2 digit ekstra di belakang koma
            'omset' : 'Rp{:.2f}'}
laporan = df.style.format(formatku)
laporan
```

```python
#PERSIAPAN DATA FRAME

d1= {'col 1' : [1,2,3],
     'col 2': [10,20,30]}
df1 = pd.DataFrame(d1)
df1
```

```python
d2= {'col 3' : [4,5,6],
     'col 4': [40,50,60]}
df2 = pd.DataFrame(d2)
df2
```

```python
#Menggabungkan (merge) 2 Data Frame secara berdampingan

df = pd.merge(df1, df2, left_index=True, right_index=True) #penggabungan berdasarkan kolom/berdampingan
df
```
