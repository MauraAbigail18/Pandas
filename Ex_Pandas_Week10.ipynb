```python
#Seleksi baris dengan banyak kriteria pada Data Frame
#PERSIAPAN DATA FRAME

df = pd.read_csv('/content/titanicfull.csv') #Sebagai Data Frame
df.head()
```

```python
df[(df['sex']=='female') & (df['age']>=60) & (df['embarked']== 'S') & (df['survived']==1)] #cara1:merangkai kriteria secara 1 baris, cocok untuk kriteria sebanyak 2-3
```

```python
df[
    (df['sex']=='female') & 
    (df['age']>=60) & 
    (df['embarked']== 'S') &
    (df['survived']==1)
                  #cara2:merangkai setiap kriteria dipisahkan dalam satu baris terpisah agar codenya dapat lebih mudah untuk dibaca
                  #cocok untuk kriteria sebanyak 4-5 atau lebih
    ]
```

```python
kr1= df['sex']=='female'
kr2= df['age']>=60
kr3= df['embarked']== 'S'
kr4= df['survived']==1
                      #cara3:memasukkan setiap kriterianya ke dalam variabel tersendiri
df[kr1&kr2&kr3&kr4] #untuk melakukan seleksi baris, cocok untuk kriteria sebanyak 4-5 atau lebih dan untuk memperjelas maksud

#ketika suatu kondisi atau kriteria kita masukkan ke dalam suatu variabel maka, sewaktu kita merangkaikannya, di setiap kriteria ini tidak perlu kita update lagi dengan penanda kurung
```

```python
#Akses sekelompok data pada Data Frame yang sudah terkelompok
#PERSIAPAN DATA FRAME

df = pd.read_csv('/content/titanicfull.csv') #Sebagai Data Frame
df.head()
```

```python
#Mengakses sekelompok data yang sudah terkelompok dengan get_group()

grouped_df = df.groupby('sex') #mengelompokkan berdasarkan sex nya
```

```python
grouped_df.get_group('female').head(10) #untuk mengakses data yang berasal dari kelompok sex berupa female sebanyak 10 data
```

```python
grouped_df = df.groupby('survived') #mengelompokkan berdasarkan survived status
```

```python
grouped_df.get_group(1).head(10) #untuk mengakses data yang berasal dari kelompok survived yang bernilai satu sebanyak 10 data
```
